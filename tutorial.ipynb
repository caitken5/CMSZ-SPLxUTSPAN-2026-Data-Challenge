### Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotly import graph_objects as go
import json
import ast
import re
```

### Import Data

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
keypoint_names = df.columns[3:-3]
print(list(keypoint_names))
```

Create function to change string arrays to numpy arrays to be able to be used effectively later on

```python
def parse_array_json(s):
    # Replace bare nan with null (JSON compatible)
    s = s.replace('nan', 'null')
    return np.array(json.loads(s), dtype=np.float32)

for col in keypoint_names:
    df[col] = df[col].apply(parse_array_json)
```

### Describe Data

```python
participants = df['participant_id'].unique()
print(f"Number of participants: {len(participants)}")
```

```python
for participant in participants:
    participant_data = df[df['participant_id'] == participant].reset_index(drop=True)
    print('=====================================')
    print(f"Participant {participant}")
    print(f"Total Shots in Training Data: {len(participant_data)}")
    print(f"Average Angle: {participant_data['angle'].mean():.2f} ± ({participant_data['angle'].std():.2f}) degrees")
    print(f"Average Depth: {participant_data['depth'].mean():.2f} ± ({participant_data['depth'].std():.2f}) in.")
    print(f"Average Left/Right: {participant_data['left_right'].mean():.2f} ± ({participant_data['left_right'].std():.2f}) in.")
```

### Visualize Data

```python
FRAME_RATE = 60
```

```python
# plot all right_wrist z positions for all player shots
fig = go.Figure()

colors = ['red', 'green','blue', 'orange',' purple']
count = 0
for participant in participants:
    participant_df = df[df['participant_id'] == participant].reset_index(drop=True)
    for i in range(len(participant_df)):
        z = participant_df[f'right_wrist_z'][i]
        t = np.arange(len(z)) / FRAME_RATE 
        fig.add_trace(go.Scatter(x=t, y=z, mode='lines',
                                    line=dict(color=colors[count], width=1),
                                    showlegend=False
                                    ))
    count += 1

fig.update_layout(
    title=f'Participant Right Wrist Z-Positions Over Time',
    xaxis_title='Time (seconds)',
    yaxis_title='Z-Coordinate (feet)',
    showlegend=False
)

# fake legend just with colors 1-5 for particpants
legend_items = []
for i, participant in enumerate(participants):
    legend_items.append(go.Scatter(x=[None], y=[None], mode='markers',
                                   name=f'Participant {participant}',
                                      marker=dict(color=colors[i], size=10)))
fig.add_traces(legend_items)

fig.update_layout(showlegend=True)


fig.show()
```

```python
# scatterplot angle to depth for all participants

fig = go.Figure()

colors = ['red', 'green','blue', 'orange',' purple']
count = 0
for participant in participants:
    participant_df = df[df['participant_id'] == participant].reset_index(drop=True)
    fig.add_trace(go.Scatter(x=participant_df['angle'], y=participant_df['depth'],
                                mode='markers',
                                marker=dict(color=colors[count], size=6, opacity=0.7),
                                name=f'Participant {participant}'
                                ))
    # add regression
    z = np.polyfit(participant_df['angle'], participant_df['depth'], 1)
    p = np.poly1d(z)
    fig.add_trace(go.Scatter(x=participant_df['angle'], y=p(participant_df['angle']),
                                mode='lines',
                                line=dict(color=colors[count], width=2, dash='dash'),
                                showlegend=False
                                ))

    count += 1

fig.update_layout(
    title='Angle vs Depth for All Participants',
    xaxis_title='Angle (degrees)',
    yaxis_title='Depth (inches)',
)
fig.show()
```

```python
# scatterplot angle to left_right for all participants

fig = go.Figure()

colors = ['red', 'green','blue', 'orange',' purple']
count = 0
for participant in participants:
    participant_df = df[df['participant_id'] == participant].reset_index(drop=True)
    fig.add_trace(go.Scatter(x=participant_df['angle'], y=participant_df['left_right'],
                                mode='markers',
                                marker=dict(color=colors[count], size=6, opacity=0.7),
                                name=f'Participant {participant}'
                                ))
    # add regression
    z = np.polyfit(participant_df['angle'], participant_df['left_right'], 1)
    p = np.poly1d(z)
    fig.add_trace(go.Scatter(x=participant_df['angle'], y=p(participant_df['angle']),
                                mode='lines',
                                line=dict(color=colors[count], width=2, dash='dash'),
                                showlegend=False
                                ))
                                
    count += 1

fig.update_layout(
    title='Angle vs Depth for All Participants',
    xaxis_title='Angle (degrees)',
    yaxis_title='Left/Right (inches)',
)
fig.show()
```
